# 1. Prepare SSH connection

```python
import pxssh
import time
import os
import shutil
import subprocess

s = pxssh.pxssh()
if not s.login ('192.168.0.200', 'root', ''):
    print "SSH session failed on login."
    print str(s)
    exit(1)
else:
    print "SSH session login successful"
    
s.sendline ('cd /tunning')
s.prompt()
print s.before

timestr = time.strftime("%Y%m%d_%H%M%S")
print timestr

dir_name = '/srv/rootfs/power_auto_test_' + timestr
print dir_name
os.mkdir(dir_name)

tmp = os.popen("ls -l /srv").read()
print tmp
```

```python
s.sendline ('')
s.prompt()         # match the prompt
print s.before     # print everything before the prompt.

os.mkdir(dir_name + "/" + "mp3/")
aep_name = dir_name + "/" + "mp3/" + "aep.txt"
print aep_name
print subprocess.Popen(["arm-probe --config /home/leoy/Work/disk/tools/x86/arm-probe/config -x -c V4.2 -a 1 -m 300 -l 20000 > " + aep_name + "&"], shell=True, stdout=subprocess.PIPE).stdout.read()

s.sendline ('./idlestat --trace -f ./result/mp3/trace.log -t 30 -p -c -w -o ./result/mp3/report.log -- ./rt-app ./doc/examples/mp3-long.json')
s.prompt()         # match the prompt
print s.before     # print everything before the prompt.

s.sendline ('mv mp3* result/mp3/')
s.prompt()         # match the prompt
print s.before     # print everything before the prompt.

time.sleep(5)
```
